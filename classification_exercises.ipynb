### Iris Data

```python
# Use the function defined in acquire.py to load the iris data.
import acquire as ac

# loading iris data using acquire from external file
iris_df = ac.new_get_iris_data()

# displaying DF
iris_df
```

```python
# Drop the species_id and measurement_id columns.

# naming 2 columns to drop 
cols_to_drop = ['species_id', 'measurement_id']

# dropping the 2 chosen columns
iris_df = iris_df.drop(columns = cols_to_drop)

# displaying DF
iris_df
```

```python
# Rename the species_name column to just species.

# renaming column
iris_df.rename(columns={'species_name': 'species'}, inplace=True)

# displaying DF
iris_df
```

```python
# Create dummy variables of the species name.

# creating dummy DF for data in species column
dummy_df = pd.get_dummies(iris_df['species'], dummy_na = False)

# adding dummy DF to original DF
iris_df = pd.concat([iris_df, dummy_df], axis=1)

# displaying DF
iris_df
```

```python
# Create a function named prep_iris that accepts the untransformed iris data, 
# and returns the data with the transformations above applied.

# creating function that will change iris DF
def prep_iris(idf):
    # naming columns to drop
    cols_to_drop = ['species_id', 'measurement_id']
    # dropping named columns
    idf = idf.drop(columns = cols_to_drop)
    # renaming columns
    idf.rename(columns={'species_name': 'species'}, inplace=True)
    # creating dummy df with data from species column
    dummy_df = pd.get_dummies(idf['species'], dummy_na = False)
    # adding dummy df to original df
    idf = pd.concat([idf, dummy_df], axis=1)
    # returning df
    return idf

# using function to test success 
prep_iris(ac.new_get_iris_data())

    
```

### TITANIC DATA

```python
# Use the function you defined in acquire.py to load the titanic data set.

# using acquire function to load frame
titanic_df = ac.new_get_titanic_data()

# displaying DF
titanic_df
```

```python
# Handle the missing values in the embark_town and embarked columns.

# naming columns to drop
titanic_df  = titanic_df.drop(columns = 'embark_town')

# dropping named columns
titanic_df  = titanic_df[~titanic_df.embarked.isnull()]

# displaying DF
titanic_df 
```

```python
# Remove the deck column.

# dropping column
titanic_df = titanic_df.drop(columns = 'deck')

# displaying DF
titanic_df 
```

```python
# Create a dummy variable of the embarked column.

# creating dummy df with data from embarked column
dummy_df = pd.get_dummies(titanic_df['embarked'], dummy_na = False)

# attaching df to original df
titanic_df = pd.concat([titanic_df, dummy_df], axis=1)

# displaying DF
titanic_df
```

```python
# Scale the age and fare columns using a min max scaler. 

# importing scaler from sklearn
from sklearn.preprocessing import MinMaxScaler

# creating variable to hold scaler function
scaler = MinMaxScaler()

# fitting and transforming data in age and far columns
titanic_df[['age', 'fare']] = scaler.fit_transform(titanic_df[['age', 'fare']])

# displaying DF
titanic_df
                 
# Why might this be beneficial?

# Answer: To handle situations where different units of measurement could lead to differences in weight. For example,
# we may have one feature measured in feet and another measured in centimeters

# When might you not want to do this?
# Answer: If the features all share the same units of measure
```

```python
# Fill the missing values in age. The way you fill these values is up to you. 
# Consider the tradeoffs of different methods.

# importing for use
from sklearn.impute import SimpleImputer
import warnings
warnings.filterwarnings("ignore")

# creating variable to hold simpleputer function with strategy set to most frequent
imputer = SimpleImputer(strategy = 'most_frequent')

# fitting imputer to the most frequent age in the age column
imputer = imputer.fit(titanic_df[['age']])

# updating all null values in age to the most frequent value in the column
titanic_df[['age']] = imputer.transform(titanic_df[['age']])

# displaying DF
titanic_df
```

```python
# Create a function named prep_titanic that accepts the untransformed titanic data, 
# and returns the data with the transformations above applied.

# prep titanic data
def prep_titanic(tdf):
    # importing for use
    from sklearn.preprocessing import MinMaxScaler
    from sklearn.impute import SimpleImputer
    import warnings
    warnings.filterwarnings("ignore")
    # dropping column
    tdf = tdf.drop(columns = 'embark_town')
    # removing rows with empty embarked value
    tdf = tdf[~tdf.embarked.isnull()]
    # dropping column
    tdf = tdf.drop(columns = 'deck')
    # creating dataframe with dummy values
    dummy_df = pd.get_dummies(tdf['embarked'], dummy_na = False)
    # adding dummy values to original data frame
    tdf = pd.concat([tdf, dummy_df], axis=1)
    # creating variable that holds minmaxscaler function
    scaler = MinMaxScaler()
    # fitting and transforming data in age and far columns
    tdf[['age', 'fare']] = scaler.fit_transform(tdf[['age', 'fare']])
    # setting variable with simple imputer set to most frequent
    imputer = SimpleImputer(strategy = 'most_frequent')
    # filling all age values that were empty with most frequent value from age column
    imputer = imputer.fit(tdf[['age']])
    tdf[['age']] = imputer.transform(tdf[['age']])
    # returning DF
    return tdf


prep_titanic(ac.new_get_titanic_data())
```

#### Is the sepal length significantly different in virginica compared to versicolor? 
#### Run a statistical experiment to test this.
#### Make sure to include a null hypothesis, alternative hypothesis, results, and summary.

```python
import scipy.stats as stats

# Hypothesis:
# H0: Virginica sepal legnths are not different than versicolor ones
# Ha: Virginica sepal legnths are different than versicolor ones

# Results
# setting alpha to represent alpha
alpha = .05

# creating 2 data frames that hold the sepal lengths for virginica and versicolor respectively
virginica_sl = train[train['species'] == 'virginica']['sepal_length']
versicolor_sl = train[train['species'] == 'versicolor']['sepal_length']

# calculating t and p value to determine if we reject or fail to reject H0
t, p = stats.ttest_ind(virginica_sl, versicolor_sl, equal_var = False)
t, p

# Summary
# We performed a 2 sample, 2 tailed test and found p to be less than alpha
# With this information, we reject the null hypothesis 
# that virginica sepal lengths are not different than versicolor ones
```

#### What is your takeaway from this statistical testing?

```python
# Answer: My takeaway is that the sepal lengths of virginicas are different than versicolors on average.
# Given that the t value was so high, this indicates that they are probably longer on average.
```
