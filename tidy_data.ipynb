#### Load the attendance.csv file and calculate an attendnace percentage for each student. One half day is worth 50% of a full day, and 10 tardies is equal to one absence.

```python
# creating DF using data from attendance fil
att = pd.read_csv('attendance.csv')

# displaying DF
att
```

```python
# replacing letter values with numbers that correspond to each attendance value
att = att.replace('P',1)
att = att.replace('H',.5)
att = att.replace('T',.9)
att = att.replace('A',0)

# displaying DF
att
```

```python
# creating series that holds the sum of the values of each row 
total = att.sum(axis = 1)

# creating series that holds the the number of columns minus the student name column (ie. the number of days)
days = att.shape[1] - 1

# creating column that holds the value for total divided by the total number days
# ie. attendance percentage
att['percentage'] = round((total/days) * 100, 2)


att
```

```python
# renaming student column
att = att.rename(columns = {'Unnamed: 0': 'student'})

# creating DF that holds student name and attendance percentage
att_final = att[['student', 'percentage']]

# displaying DF
att_final
```

#### Read the cake_recipes.csv data. This data set contains cake taste_score scores for combinations of different recipes, oven rack positions, and oven temperatures.

```python
# creating DF using data from cake recipes file
cake = pd.read_csv('cake_recipes.csv')

# displaying DF
cake.head()
```

#### Tidy the data as necessary.

```python
# renaming column in preparation of split
cake = cake.rename(columns={'recipe:position':'position'})

# spliting recipe and position into separate columns
recipe_position_s = cake.position.str.split(':', expand=True)

# creating DF that only includes recipe and oven position columns
recipe_position_s.columns = ['recipe', 'oven_position']

# adding recipe/position DF to original DF
new_cake = pd.concat([cake, recipe_position_s], axis=1)

# dropping position column because we already its data in another column
new_cake = new_cake.drop(['position'],axis=1)

# using melt to make temperature its own column while creating a new column "taste_score" which holds
# the taste value for each combo of oven position, recipe, and temp ie. each row
melt_cake = new_cake.melt(id_vars=['recipe', 'oven_position'], var_name='temp', value_name='taste_score')

# reviewing results
melt_cake

```

#### Which recipe, on average, is the best? recipe b

```python
# grouping by recipe and displaying the average taste score each recipe
melt_cake.groupby('recipe').taste_score.mean()

# Recipe B has the highest average so its the best recipe on average
```

#### Which oven temperature, on average, produces the best results? 275

```python
# grouping by temp and displaying the average taste score per temp
melt_cake.groupby('temp').taste_score.mean()

# Temp 275 produces the highest taste score on average
```

#### Which combination of recipe, rack position, and temperature gives the best result? 
#### recipe b, bottom rack, 300 degrees

```python
# sorting df by taste_score so that the highest score is at the bottom of the df
# using tail(1) to display to last row of the sorted df, ie. the one with highest score
melt_cake.sort_values(by='taste_score').tail(1)

# Answer: Recipe b, bottom position, 300 temp
```
