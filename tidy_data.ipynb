#### Load the attendance.csv file and calculate an attendnace percentage for each student. One half day is worth 50% of a full day, and 10 tardies is equal to one absence.

```python
# creating DF using data from attendance fil
att = pd.read_csv('attendance.csv')

# displaying DF
att
```

```python
# replacing letter values with numbers that correspond to each attendance value
att = att.replace('P',1)
att = att.replace('H',.5)
att = att.replace('T',.9)
att = att.replace('A',0)

# displaying DF
att
```

```python
# creating series that holds the sum of the values of each row 
total = att.sum(axis = 1)

# creating series that holds the the number of columns minus the student name column (ie. the number of days)
days = att.shape[1] - 1

# creating column that holds the value for total divided by the total number days
# ie. attendance percentage
att['percentage'] = round((total/days) * 100, 2)


att
```

```python
# renaming student column
att = att.rename(columns = {'Unnamed: 0': 'student'})

# creating DF that holds student name and attendance percentage
att_final = att[['student', 'percentage']]

# displaying DF
att_final
```
